```python
%matplotlib notebook
import numpy as np
from matplotlib import pyplot as plt
from ipywidgets import interact
from packages.utils.videos import VideoSource
```

```python
x = np.arange(0, 2*np.pi)
vs = VideoSource('./videos/example_01.mp4')
```

```python
fig1, axes1 = plt.subplots(figsize=(8,6))

def update_plot(T=0.1):
    axes1.cla()
    axes1.plot(x, np.sin(x*T))
interact(update_plot, T=(0, 1, 0.1))
```

```python
fig2, axes2 = plt.subplots()

def update_plot(T=0.1):
    axes2.cla()
    axes2.plot(x, np.sin(2*x*T))
interact(update_plot, T=(2.0, 4.0, 0.1))
```

```python
fig3, axes3 = plt.subplots()

def update_image(f=0):
    axes3.cla()
    axes3.imshow(vs.get_frame(f))


interact(update_image, f=(0, vs.get_total_frames() - 1, 1))
```
